```python
import pdfplumber
import pandas as pd
import os
# import re
fileList = os.listdir()
fileList = [i for i in os.listdir() if i[-3:] == 'pdf'or i[-3:] == 'PDF']

for i in fileList:
    with pdfplumber.open(i) as pdf:
        page_count = len(pdf.pages)
        f = open(i.replace('PDF','txt'), 'w',encoding = 'utf-8')
        for page in pdf.pages:
            results = page.extract_words()
            print(results)
            
            for  word in results:
                f.write(word['text']+'\n')

            # print(pdf_table)
        f.close()
```
